```python
big_list = []
# Loop through each item in the list
for item in entry_list:
    number_hits = 0
    link = "https://prosite.expasy.org/cgi-bin/prosite/scanprosite/PSScan.cgi?seq=" + item
    response = requests.get(link)
    save = response.text
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(save, 'html.parser')
    
    # Find the <pre> tag and get its text content
    pre_content = soup.find('pre').get_text()

    # Split the text into lines
    lines = pre_content.split('\n')

    # Extract the line that starts with "found:"
    found_line = next((line for line in lines if line.startswith("found:")), None)
    
    # Extract the number from the found line
    if found_line:
        found_number = int(found_line.split()[1])
    else:
        found_number = None

    variables = []

    if found_number !=0 :

        if found_number is None:
            variables = [item, '','','','','']
            comment = 'noref_proteome'

        elif found_number <= 1:
            # Extract lines that contain '>'
            extracted_lines = [line for line in lines if '>' in line]
            line = extracted_lines[0]
            
            # Split line by '|' and take elements after '|'
            parts = line.split('|')[1:]
            for part in parts:
                # Clean part and save to list
                comment = 'found'
                cleaned_part = part.split()[0]
                variables.append(cleaned_part)
                #reorder list
                first_item = variables.pop(0)  # Remove and get the first item
                variables = [item, '', first_item, '','','']

        else:
            # Extract lines that contain '>'
            extracted_lines = [line for line in lines if '>' in line]
        
            for line in extracted_lines:
                # Split line by '|' and take elements after '|'
                parts = line.split('|')[1:]
                for part in parts:
                    # Clean part and save to list
                    comment = 'found'
                    cleaned_part = part.split()[0]
                    variables.append(cleaned_part)
                number_hits = number_hits + 1

            for _ in range(number_hits):  # Loop 'number_hits' times
                variables.append('')
                
    else:
        variables = [item, '','','','','']
        comment = "no hit!"
    variables.append(comment)
    big_list.append(variables)
```

```python
def apply_conditions(row):
    if row['Hit1'] == 'GAIN_B':
        if row['Hit2'] == 'F2_4':
            return 'Sec'
    if row['Hit1'] == 'F1_2':
        if row['Hit2'] == 'F2_4':
            return 'cAMP'
        else:
            return 'Rho'
    if row['Hit1'] == 'F2_4':
        if row['Hit2'] == 'F1_2':
            if row['Hit3'] == 'PIPK':
                return 'Git3-PIPK'
            else:
                return 'Git3'
        if row['Hit2'] == 'PIPK':
            return 'cAMP-PIPK'
        else:
            return 'cAMP'
    elif (row['Hit1'] == 'F3_4') or (row['Hit2'] == 'F3_4'): ### check if this code is ok 
        return 'Glu'
    elif row['Hit1'] == 'F2_3':
        if row['Hit2'] == 'F2_4':
            return 'Sec'
        else:
            return 'Sec-putative'
    elif row['Hit1'] == 'FZ':
        return 'Friz'
    else:
        return 'NN'

# Apply the function to each row in the DataFrame to create the new column
df_mod['Expasy Classification'] = df_mod.apply(apply_conditions, axis=1)
```

```python
with open(filtered_entries, 'r') as file_a:
    list_a = file_a.readlines()

with open(misannotations, 'r') as file_b:
    list_b = file_b.readlines()

# Remove newline characters for accurate comparison
list_a = [line.strip() for line in list_a]
list_b = [line.strip() for line in list_b]

# Filter out the entries in list_a that are present in list_b
filtered_list = [item for item in list_a if item not in list_b]

# Optionally, write the filtered list back to file_a or a new file
with open(new_output, 'w') as output_file:
    for item in filtered_list:
        output_file.write(item + '\n')

print("Filtered content has been saved to 'expassy_filtered.txt'")
```
