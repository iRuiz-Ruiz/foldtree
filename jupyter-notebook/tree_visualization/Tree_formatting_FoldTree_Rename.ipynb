```python
#Import nexus template
# Specify the path to your NEXUS file
midblock_filename = '../../tree_blocks/midblock.txt'

# Read the file as text
with open(midblock_filename, 'r') as f:
    midblock = f.read()

# Print or process the text
print(midblock)
```

```python
# Replace 'count' with upblock input_value
midblock_tree = midblock.replace('include_tree_here', str(input))
print(midblock_tree)
```

```python
# Create a dictionary for replacement
replace_dict = dict(zip(df['Entry'], df['full_label']))

replacements =[]

# Function to replace occurrences
def replace_entry(match):
    entry = match.group(1)
    replaced_entry = replace_dict.get(entry, entry)
    
    # Append the original and replaced values to the list if they were replaced
    if entry != replaced_entry:
        replacements.append((entry, replaced_entry))
    
    return replaced_entry
    
# Perform replacement using regular expressions
pattern = r'([A-Za-z0-9_]+)'
nexus_text_replaced = re.sub(pattern, replace_entry, midblock_tree)

# Convert the list of replacements to a DataFrame
replacements_df = pd.DataFrame(replacements, columns=['Original', 'Replaced'])

# Print modified text
print(nexus_text_replaced)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

# Define the custom colormap as a gradient
colors = ["yellow", "red", "purple"]
n_bins = 100  # Number of bins for the gradient
custom_colormap = LinearSegmentedColormap.from_list("custom_colormap", colors, N=n_bins)

# Pivot the DataFrame
pivot_table = tax_sum2.pivot(index="Label", columns=2, values="count")

# Create a mask for missing values
mask = pivot_table.isnull()

# Create a heatmap
sns.heatmap(data=pivot_table, annot=True, cmap=custom_colormap, mask=mask, cbar_kws={'label': 'Count'})

# Display the plot
plt.show()
```
